```python
#Sütunlara ait boş hücre sayısını bulmak için

bos_deger_sayisi = veri.isnull().sum()
bos_deger_sayisi
```

```python
#Datamızın bir kısmı ile çalışak istersek

ders_verileri = veri.loc[ :, "turkce_dogru":"toplam_net"]
ders_verileri.head()
```

```python
#Tüm boş değerleri 0 ile değiştirmek için
# replace all NA's with 0
ders_verileri.fillna(0).head()
```

```python
#Boş değeri sütunda kendisinden sonra gelen değerle doldurmak için
ders_verileri.fillna(method='bfill', axis=0)
```

```python
# 1. NaN değerlerini kendisinden hemen sonra gelen değerle doldur (method='bfill')
# 2. Hala NaN kalan yerleri ( en sonda olup arkasında veri olmayanlar) 0 ile doldur
veri.fillna(method='bfill', axis=0).fillna(0)
```

```python
#Sütunda bulunan benzersiz değerleri öğrenmek için

subeler = veri['sube'].unique()
subeler
```

```python
#Sütunda bulunan değerin önünde ve arkasındaki boşlukları kaldırmak için

veri['sube'] = veri['sube'].str.strip()
veri['sube'].unique()
```

```python
#istemediğimiz bir şuba adını düzelmek için 0 ı D yapmak için

veri['sube'] = veri['sube'].replace('0', 'D')

veri['sube'].unique()
```

```python
# Şube sütunu '0' olmayan satırları tut (yani '0' olanları sil)
veri = veri[veri['sube'] != '0']

# Kontrol için
print(veri['sube'].unique())
```

```python
#istenilen sütundaki değerleri küçük harfe dönüştürmek isteseydim

veri['sube'] = veri['sube'].str.lower()
veri['sube'].unique()
```
