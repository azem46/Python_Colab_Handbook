# **Python JSON**

JSON, veri depolama ve alışverişi için kullanılan bir sözdizimidir.

JSON, JavaScript nesne gösterimiyle yazılmış metindir.

JSON modülünü içe aktarın:

import json

```python
#JSON'dan Python'a dönüştürme:

import json

# some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'

# parse x:
y = json.loads(x)

# the result is a Python dictionary:
print(y["age"])
```

```python
#Eğer bir Python nesneniz varsa, bu nesneyi bir JSON dizesine dönüştürmek için ilgili yöntemi kullanabilirsiniz json.dumps().

import json

# a Python object (dict):
x = {
  "name": "John",
  "age": 30,
  "city": "New York"
}

# convert into JSON:
y = json.dumps(x)

# the result is a JSON string:
print(y)
```

```python
#Python nesnelerini JSON dizelerine dönüştürün ve değerleri yazdırın:

import json

print(json.dumps({"name": "John", "age": 30}))
print(json.dumps(["apple", "bananas"]))
print(json.dumps(("apple", "bananas")))
print(json.dumps("hello"))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

```python
#Geçerli tüm veri türlerini içeren bir Python nesnesini dönüştürün:

import json

x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

print(json.dumps(x))
```

```python
#indentGirinti sayısını tanımlamak için parametreyi kullanın :

json.dumps(x, indent=4)
```

```python
#separatorsVarsayılan ayırıcıyı değiştirmek için bu parametreyi kullanın :

json.dumps(x, indent=4, separators=(". ", " = "))
```

```python
#sort_keysSonucun sıralanıp sıralanmayacağını belirtmek için bu parametreyi kullanın :

json.dumps(x, indent=4, sort_keys=True)
```

# **Python RegEx**

Düzenli ifade (RegEx), bir arama deseni oluşturan karakter dizisidir.

Düzenli ifadeler (regex), bir dizenin belirtilen arama desenini içerip içermediğini kontrol etmek için kullanılabilir.

```python
#Aşağıdaki metin dizisinde "The" ile başlayıp "Spain" ile biten ifadeler olup olmadığını kontrol edin:

import re

txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)
print(x)
```

```python
#Tüm maçların listesini yazdırın:

import re

txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)
```

```python
#Eşleşme bulunamazsa boş bir liste döndürün:

import re

txt = "The rain in Spain"
x = re.findall("Portugal", txt)
print(x)
```

```python
#Dizedeki ilk boşluk karakterini arayın:

import re

txt = "The rain in Spain"
x = re.search("\s", txt)

print("The first white-space character is located in position:", x.start())
```

```python
#Hiçbir eşleşme bulmayan bir arama yapın:

import re

txt = "The rain in Spain"
x = re.search("Portugal", txt)
print(x)
```

```python
#Her boşluk karakterinde bölünür:
import re

txt = "The rain in Spain"
x = re.split("\s", txt)
print(x)
```

```python
#Dizeyi yalnızca ilk geçtiği yerde böl:

import re

txt = "The rain in Spain"
x = re.split("\s", txt, 1)
print(x)
```

```python
#Tüm boşluk karakterlerini 9 rakamıyla değiştirin:

import re

txt = "The rain in Spain"
x = re.sub("\s", "9", txt)
print(x)
```

```python
#İlk 2 örneği değiştirin:

import re

txt = "The rain in Spain"
x = re.sub("\s", "9", txt, 2)
print(x)
```

```python
#Eşleşen bir nesne döndürecek bir arama yapın:

import re

txt = "The rain in Spain"
x = re.search("ai", txt)
print(x) #this will print an object
```

```python
#İlk eşleşmenin konumunu (başlangıç ​​ve bitiş konumunu) yazdırın.

#Bu düzenli ifade, büyük harf "S" ile başlayan tüm kelimeleri arar:

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span())
```

```python
#Fonksiyona iletilen metni yazdırın:

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.string)
```

```python
#Eşleşme bulunan dize bölümünü yazdırın.

#Bu düzenli ifade, büyük harf "S" ile başlayan tüm kelimeleri arar:

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.group())
```

```python
import re

txt = "That will be 59 dollars"

#Find all digit characters:

x = re.findall("\d", txt)
print(x)
```

Metakarakterler
Metakarakterler, özel bir anlamı olan karakterlerdir:

Character	Description	Example	Try it
[]	A set of characters	"[a-m]"
\	Signals a special sequence (can also be used to escape special characters)	"\d"

$	Ends with	"planet$"

+	One or more occurrences	"he.+o"

{}	Exactly the specified number of occurrences	"he.{2}o"
|	Either or	"falls|stays"

re.NOFLAG		Specifies that no flag is set for this pattern
re.UNICODE	re.U	Returns Unicode matches. This is default from Python 3. For Python 2: use this flag to return only Unicode matches
re.VERBOSE	re.X	Allows whitespaces and comments inside patterns. Makes the pattern more readable

# **String Formatting**

F-string, bir dizenin seçili bölümlerini biçimlendirmenizi sağlar.

Bir dizeyi f-dizesi olarak belirtmek için, fdize sabitinin önüne bir işaret koymanız yeterlidir, örneğin:

```python
#Bir f-string oluşturun:

txt = f"The price is 49 dollars"
print(txt)
```

```python
#Değişken için bir yer tutucu ekleyin price:

price = 59
txt = f"The price is {price} dollars"
print(txt)
```

```python
#Fiyatı 2 ondalık basamaklı olarak gösterin:

price = 59
txt = f"The price is {price:.2f} dollars"
print(txt)
```

```python
#Değeri 952 ondalık basamakla gösterin:

txt = f"The price is {95:.2f} dollars"
print(txt)
```

```python
#Yer tutucuda matematiksel bir işlem gerçekleştirin ve sonucu döndürün:

txt = f"The price is {20 * 59} dollars"
print(txt)
```

```python
#Fiyatı göstermeden önce vergileri ekleyin:

price = 59
tax = 0.25
txt = f"The price is {price + (price * tax)} dollars"
print(txt)
```

```python
#Fiyat 50'nin üzerindeyse "Pahalı", aksi takdirde "Ucuz" yazın:

price = 49
txt = f"It is very {'Expensive' if price>50 else 'Cheap'}"

print(txt)
```

```python
#upper()Bir değeri büyük harflere dönüştürmek için `string` yöntemini kullanın :

fruit = "apples"
txt = f"I love {fruit.upper()}"
print(txt)
```

```python
#Ayak birimini metreye dönüştüren bir fonksiyon oluşturun:

def myconverter(x):
  return x * 0.3048

txt = f"The plane is flying at a {myconverter(30000)} meter altitude"
print(txt)
```

```python
#Binlik ayırıcı olarak virgül kullanın:

price = 59000
txt = f"The price is {price:,} dollars"
print(txt)
```

```python
txt = f"We have {49:>8} chickens."
print(txt)
```

```python
#Fiyatı göstermek istediğiniz yere bir yer tutucu ekleyin:

price = 49
txt = "The price is {} dollars"
print(txt.format(price))
```

```python
#Fiyatın iki ondalık basamaklı bir sayı olarak görüntülenmesini sağlayın:

txt = "The price is {:.2f} dollars"
```

```python
#Ve daha fazla yer tutucu ekleyin:


quantity = 3
itemno = 567
price = 49
myorder = "I want {} pieces of item number {} for {:.2f} dollars."
print(myorder.format(quantity, itemno, price))
```

```python
#{0}Değerlerin doğru yer tutuculara yerleştirildiğinden emin olmak için indeks numaralarını (süslü parantez içindeki bir sayı) kullanabilirsiniz :

quantity = 3
itemno = 567
price = 49
myorder = "I want {0} pieces of item number {1} for {2:.2f} dollars."
print(myorder.format(quantity, itemno, price))
```

```python
#Ayrıca, aynı değere birden fazla kez başvurmak istiyorsanız, indeks numarasını kullanın:

age = 36
name = "John"
txt = "His name is {1}. {1} is {0} years old."
print(txt.format(age, name))
```

```python
myorder = "I have a {carname}, it is a {model}."
print(myorder.format(carname = "Ford", model = "Mustang"))
```
