# **Matplotlib Dağılım Grafiği**

scatter()Pyplot ile dağılım grafiği çizmek için fonksiyonunu kullanabilirsiniz .

Bu scatter()fonksiyon, her gözlem için bir nokta çizer. X eksenindeki değerler için bir ve Y eksenindeki değerler için bir olmak üzere aynı uzunlukta iki diziye ihtiyaç duyar:

```python
#Basit bir dağılım grafiği:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

Yukarıdaki örnekteki gözlem, 13 aracın geçmesi sonucu elde edilmiştir.

X ekseni, aracın yaşını gösteriyor.

Y ekseni, aracın geçerkenki hızını gösterir.

Gözlemler arasında herhangi bir ilişki var mı?

Görünüşe göre araba ne kadar yeniyse o kadar hızlı gidiyor, ama bu bir tesadüf de olabilir, sonuçta sadece 13 araba kaydettik.

**Grafikleri Karşılaştır**

Yukarıdaki örnekte hız ve yaş arasında bir ilişki olduğu görülüyor, peki ya başka bir güne ait gözlemleri de grafiğe eklersek? Dağılım grafiği bize başka bir şey söyleyecek mi?

```python
#Aynı şekil üzerinde iki grafik çizin:

import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

**Renkler**

Her bir dağılım grafiği için kendi renginizi veya argümanını kullanarak belirleyebilirsiniz c :

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

**Her noktayı renklendirin**

Hatta argüman değeri olarak bir renk dizisi kullanarak her nokta için belirli bir renk bile ayarlayabilirsiniz c:

Not: Bunun için yalnızca argümanı kullanabilirsiniz, tek argümanı kullanabilirsiniz .colorc

```python
#İşaretleyicilerin rengini kendiniz belirleyin:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

# **Renk Haritası**

Matplotlib modülü, çeşitli renk haritaları sunmaktadır.

Renk haritası, her rengin 0 ile 100 arasında bir değere sahip olduğu bir renk listesi gibidir.

Renk Haritası Nasıl Kullanılır?

Renk haritasını, bu durumda Matplotlib'de bulunan yerleşik renk haritalarından biri olan cmaprenk haritasının değeriyle birlikte anahtar kelime argümanını kullanarak belirtebilirsiniz .'viridis'

Ek olarak, dağılım grafiğindeki her nokta için bir değer olmak üzere (0'dan 100'e kadar) değerlerden oluşan bir dizi oluşturmanız gerekiyor:

```python
#Gerçek renk haritasını ekleyin:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

# **Boyut**

Noktaların boyutunu argüman kullanarak değiştirebilirsiniz s.

Renklerde olduğu gibi, boyutlar için kullanılan dizinin uzunluğunun x ve y eksenleri için kullanılan dizilerle aynı olduğundan emin olun:

```python
#İşaretleyiciler için kendi boyutunuzu ayarlayın:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

# **Alfa**

Noktaların şeffaflığını argüman kullanarak ayarlayabilirsiniz alpha.

Renklerde olduğu gibi, boyutlar için kullanılan dizinin uzunluğunun x ve y eksenleri için kullanılan dizilerle aynı olduğundan emin olun:

```python
#İşaretleyiciler için kendi boyutunuzu ayarlayın:

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

```python
#Renk, Boyut ve Alfa'yı Birleştir
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# **Bar Grafik  Oluşturma**

bar()Pyplot ile çubuk grafikler çizmek için şu fonksiyonu kullanabilirsiniz :

```python
#4 çubuk çizin:

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["5-A", "5-B", "5-C", "5-D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
#Bu bar()fonksiyon, çubukların düzenini tanımlayan argümanlar alır.
#Birinci ve ikinci argümanla temsil edilen kategoriler ve değerleri diziler halinde sunulmuştur.

x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

**Yatay Çubuklar**

Çubukların dikey yerine yatay olarak görüntülenmesini istiyorsanız, şu barh()işlevi kullanın:

```python
#4 adet yatay çubuk çizin:

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

**Çubuk Rengi**

"The bar()and" ifadesi, çubukların rengini ayarlamak için barh()anahtar kelime argümanını kullanır

```python
# 140 renk kodyla 4 tane kırmızı çubuk çizin:

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

```python
# hex koduyla Güzel yeşil renkte 4 çubuk çizin:

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#4CAF50")
plt.show()
```

**Çubuk Genişliği**

Bu fonksiyon, çubukların genişliğini ayarlamak için bar()anahtar kelime argümanını kullanır :width

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

**Bar Yüksekliği**

Bu fonksiyon, çubukların yüksekliğini ayarlamak için barh()anahtar kelime argümanını kullanır :height

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

# **Matplotlib Pasta Grafikleri**

pie()Pyplot ile pasta grafikleri çizmek için şu fonksiyonu kullanabilirsiniz :

```python
#Gördüğünüz gibi, pasta grafiği dizideki her değer için bir dilim (bu durumda [35, 25, 25, 15]) çiziyor.

#Varsayılan olarak, ilk dilimin çizimi x ekseninden başlar ve saat yönünün tersine doğru ilerler :

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

**Etiketler**

Parametreyi kullanarak pasta grafiğine etiketler ekleyin labels.

Parametre labels, her bir dilim için bir etiket içeren bir dizi olmalıdır:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

**Başlangıç ​​Açısı**

Daha önce de belirtildiği gibi, varsayılan başlangıç ​​açısı x eksenindedir, ancak bir startangleparametre belirterek başlangıç ​​açısını değiştirebilirsiniz.

Parametre startangle, derece cinsinden bir açı ile tanımlanır; varsayılan açı 0'dır:

```python
#İlk kama parçasını 90 derecelik açıyla yerleştirin:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

**Öne Çıkar**

Belki de dilimlerden birinin öne çıkmasını istersiniz? Bu explodeparametre bunu yapmanıza olanak tanır.

explodeBelirtilmişse ve boş bırakılmamışsa, parametre her Nonedilim için bir değer içeren bir dizi olmalıdır.

Her değer, her bir dilimin merkezden ne kadar uzakta gösterildiğini temsil eder:

```python
#"Apple" dilimini turtanın merkezinden 0,2 mm kadar çekin:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
```

# **Gölge**

Pasta grafiğine gölge eklemek için parametreyi shadowsşu şekilde ayarlayın True:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
#Her bir dilim için yeni bir renk belirtin:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

# **Açıklama**

Her dilim için açıklama listesi eklemek için şu legend()işlevi kullanın:

```python
#Açıklama ekle:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
#Başlık içeren bir açıklama ekleyin:

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
