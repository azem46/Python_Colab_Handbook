# **Python Dosya Açma**

Dosya işlemleri, herhangi bir web uygulamasının önemli bir parçasıdır.

Python'da dosya oluşturma, okuma, güncelleme ve silme işlemleri için çeşitli fonksiyonlar bulunmaktadır.
Python'da dosyalarla çalışmanın temel işlevi, `file` fonksiyonudur open().

Bu open()fonksiyon iki parametre alır: dosya adı ve mod .

Dosya açmak için dört farklı yöntem (mod) vardır:

"r"- Oku - Varsayılan değer. Dosyayı okumak için açar, dosya yoksa hata verir.

"a"- Ekle - Ekleme yapılacak bir dosyayı açar, dosya yoksa oluşturur.

"w"- Yaz - Yazmak için bir dosya açar, dosya yoksa oluşturur.

"x"- Oluştur - Belirtilen dosyayı oluşturur, dosya zaten varsa hata döndürür.

Ek olarak, dosyanın ikili mi yoksa metin modunda mı işleneceğini belirtebilirsiniz.

"t"- Metin - Varsayılan değer. Metin modu

"b"- İkili - İkili mod (örneğin resimler)

# **Yeni Dosya Oluştur**
Python'da yeni bir dosya oluşturmak için open()aşağıdaki parametrelerden birini kullanarak yöntemi kullanabilirsiniz:

"x"- Oluştur - bir dosya oluşturur, dosya zaten varsa hata döndürür.

"a"- Ekle - Belirtilen dosya mevcut değilse yeni bir dosya oluşturur.

"w"- Yazma - Belirtilen dosya mevcut değilse dosya oluşturur.

```python
#"myfile.txt" adında yeni bir dosya oluşturun:

f = open("myfile.txt", "x")
```

```python
#Yaz ve oku
with open("myfile.txt", "a") as f:
  f.write("Now the file has more content!")

#open and read the file after the appending:
with open("myfile.txt") as f:
  print(f.read())
```

```python
with open("demofile.txt", "w") as f:
  f.write("Woops! I have deleted the content!")

#open and read the file after the overwriting:
with open("demofile.txt") as f:
  print(f.read())
```

```python
#Bir dosyayı okumak için açmak için dosya adını belirtmek yeterlidir:

f = open("myfile.txt")
```

```python
#Yukarıdaki kod, aşağıdakiyle aynıdır:

f = open("myfile.txt", "rt")
```

```python
#Dosyayı Silmel için

#"demofile.txt" dosyasını silin:

import os
os.remove("myfile.txt")
os.remove("demofile.txt")
```

```python
#Dosyanın var olup olmadığını kontrol edin, varsa silin:

import os
if os.path.exists("demofile.txt"):
  os.remove("demofile.txt")
else:
  print("The file does not exist")
```
