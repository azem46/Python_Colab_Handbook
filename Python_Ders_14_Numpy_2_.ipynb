# **Rastgele Sayı Üret**

randomNumPy, rastgele sayılarla çalışmak için bir modül sunmaktadır .

```python
#0 ile 100 arasında rastgele bir tamsayı üretin:

from numpy import random

x = random.randint(100)

print(x)
```

```python
#0 ile 1 arasında rastgele bir ondalık sayı üretin:

from numpy import random

x = random.rand()

print(x)
```

```python
#0 ile 100 arasında rastgele 5 tamsayı içeren tek boyutlu bir dizi oluşturun:

from numpy import random

x=random.randint(100, size=(5))

print(x)
```

```python
#5 adet rastgele ondalık sayı içeren 1 boyutlu bir dizi oluşturun:

from numpy import random

x = random.rand(5)

print(x)
```

```python
#Her satırında 5 rastgele sayı bulunan 3 satırlı 2 boyutlu bir dizi oluşturun:

from numpy import random

x = random.rand(3, 5)

print(x)
```

```python
#Dizideki değerlerden birini döndürün:

from numpy import random

x = random.choice([3, 5, 7, 9])

print(x)
```

```python
#Dizi parametresindeki değerlerden (3, 5, 7 ve 9) oluşan 2 boyutlu bir dizi oluşturun:

from numpy import random

x = random.choice([3, 5, 7, 9], size=(3, 5))

print(x)
```

```python
#Her birinin değeri 3, 5, 7 veya 9 olan 100 adet değer içeren tek boyutlu bir dizi oluşturun.

#Değerin 3 olma olasılığı 0,1 olarak belirlenmiştir.

#Değerin 5 olma olasılığı 0,3 olarak belirlenmiştir.

#Değerin 7 olma olasılığı 0,6 olarak belirlenmiştir.

#Değerin 9 olma olasılığı 0 olarak belirlenmiştir.

from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)
```

```python
#Yukarıdaki örneğin aynısı, ancak her biri 5 değer içeren 3 satırdan oluşan 2 boyutlu bir dizi döndürün.

from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x)
```

```python
#Aşağıdaki dizinin elemanlarını rastgele karıştırın:

from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

random.shuffle(arr)

print(arr)
```

```python
#Aşağıdaki dizinin elemanlarından rastgele bir permütasyon oluşturun:

from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print(random.permutation(arr))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot([0, 1, 2, 3, 4, 5], kind="kde")

plt.show()
```

```python
#2x3 boyutunda rastgele bir normal dağılım oluşturun:

from numpy import random

x = random.normal(size=(2, 3))

print(x)
```

```python
#Ortalaması 1 ve standart sapması 2 olan 2x3 boyutunda rastgele bir normal dağılım oluşturun:

from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(size=1000), kind="kde")

plt.show()
```

```python
#Madeni paranın 10 kez atılmasıyla 10 veri noktası üretin:

from numpy import random

x = random.binomial(n=10, p=0.5, size=10)

print(x)
```

```python
#Binom Dağılımının Görselleştirilmesi
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n=10, p=0.5, size=1000))

plt.show()
```

Normal ve Binom Dağılımı Arasındaki Fark
Temel fark, normal dağılımın sürekli, binom dağılımının ise kesikli olmasıdır; ancak yeterli veri noktası varsa, belirli bir konum ve ölçekte normal dağılıma oldukça benzer olacaktır.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(loc=50, scale=5, size=1000),
  "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

# **Poisson Dağılımı**

Poisson dağılımı ayrık bir dağılımdır .

Belirli bir zaman diliminde bir olayın kaç kez gerçekleşebileceğini tahmin eder. Örneğin, bir kişi günde iki kez yemek yiyorsa, üç kez yemek yeme olasılığı nedir?

İki parametresi vardır:

lam- Oran veya bilinen oluşum sayısı, örneğin yukarıdaki problem için 2.

size- Döndürülen dizinin şekli.

```python
#2. olay için rastgele bir 1x10 dağılımı oluşturun:

from numpy import random

x = random.poisson(lam=2, size=10)

print(x)
```

```python
#Poisson Dağılımının Görselleştirilmesi

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.poisson(lam=2, size=1000))

plt.show()
```

Normal ve Poisson Dağılımları Arasındaki Fark

Normal dağılım sürekli iken Poisson dağılımı kesiklidir.

Ancak, yeterince büyük bir Poisson dağılımı için, binom dağılımına benzer şekilde, belirli bir standart sapma ve ortalama ile normal dağılıma benzer hale geleceğini görebiliriz.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(loc=50, scale=7, size=1000),
  "poisson": random.poisson(lam=50, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

Binom ve Poisson Dağılımları Arasındaki Fark

Binom dağılımının yalnızca iki olası sonucu varken, Poisson dağılımının sınırsız sayıda olası sonucu olabilir.

Ancak çok büyük nve sıfıra yakın değerler için pbinom dağılımı, Poisson dağılımına neredeyse özdeştir, öyle ki n * pyaklaşık olarak eşittir lam.

```python
#Binom ve Poisson Dağılımları Arasındaki Fark
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "binomial": random.binomial(n=1000, p=0.01, size=1000),
  "poisson": random.poisson(lam=10, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

Tekdüze Dağıtım

Her olayın gerçekleşme olasılığının eşit olduğu durumlarda olasılığı tanımlamak için kullanılır.

Örneğin, rastgele sayı üretimi.

Üç parametresi vardır:

low- alt sınır - varsayılan değer 0,0.

high- üst sınır - varsayılan değer 1.0.

size- Döndürülen dizinin şekli.

```python
#2x3 boyutunda tekdüze dağılımlı bir örneklem oluşturun:

from numpy import random

x = random.uniform(size=(2, 3))

print(x)
```

```python
#Tekdüze Dağılımın Görselleştirilmesi
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.uniform(size=1000), kind="kde")

plt.show()
```

Lojistik Dağıtım

Lojistik dağıtım, büyümeyi tanımlamak için kullanılır.

Lojistik regresyon, sinir ağları vb. makine öğrenmesinde yaygın olarak kullanılır.

Üç parametresi vardır:

loc- Tepe noktasının ortalama değeri. Varsayılan değer 0.

scale- Standart sapma, dağılımın düzlüğü. Varsayılan değer 1.

size- Döndürülen dizinin şekli.

```python
#Ortalaması 1 ve standart sapması 2,0 olan lojistik dağılımdan 2x3 örneklem alın:

from numpy import random

x = random.logistic(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
#Lojistik Dağıtımın Görselleştirilmesi
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.logistic(size=1000), kind="kde")

plt.show()
```

Lojistik ve Normal Dağılım Arasındaki Fark

Her iki dağılım da neredeyse aynıdır, ancak lojistik dağılımın kuyruklarının altında daha fazla alan bulunur; bu da ortalamadan daha uzakta bir olayın meydana gelme olasılığının daha yüksek olduğu anlamına gelir.

Ölçek değeri (standart sapma) ne kadar yüksekse, normal ve lojistik dağılımlar tepe noktası dışında neredeyse aynıdır.

```python

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(scale=2, size=1000),
  "logistic": random.logistic(size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

# **Çoklu Dağılım**

Çoklu dağılım, ikili dağılımın bir genellemesidir.

İki terimli dağılımdan farklı olarak, çok terimli senaryoların sonuçlarını tanımlar; iki terimli dağılımda senaryolar yalnızca iki seçenekten biri olmalıdır. Örneğin, bir popülasyonun kan grubu, zar atma sonucu.

Üç parametresi vardır:

n- Deneyin kaç kez çalıştırılacağı.

pvals- Olası sonuçların listesi (örneğin, zar atışı için [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]).

size- Döndürülen dizinin şekli.

```python
#Zar atışı için bir örnek çizin:

from numpy import random

x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(x)
```

# **Üstel Dağılım**

Üstel dağılım, bir sonraki olaya kadar geçen süreyi (örneğin başarısızlık/başarı vb.) tanımlamak için kullanılır.

İki parametresi vardır:

scale- Oranın tersi (Poisson dağılımındaki lam'a bakın) varsayılan olarak 1,0'dır.

size- Döndürülen dizinin şekli.

```python
#2x3 boyutunda ve 2,0 ölçekli üstel dağılım için bir örneklem çizin:

from numpy import random

x = random.exponential(scale=2, size=(2, 3))

print(x)
```

```python
#Üstel Dağılımın Görselleştirilmesi

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.exponential(size=1000), kind="kde")

plt.show()
```

Poisson ve Üstel Dağılımlar Arasındaki İlişki

Poisson dağılımı, belirli bir zaman diliminde bir olayın meydana gelme sayısıyla ilgilenirken, üstel dağılım bu olaylar arasındaki zamanla ilgilenir.

# **Ki Kare Dağılımı**

Ki kare dağılımı, hipotezi doğrulamak için temel olarak kullanılır.

İki parametresi vardır:

df- (serbestlik derecesi).

size- Döndürülen dizinin şekli.

```python
#Serbestlik derecesi 2 olan ki kare dağılımı için 2x3 boyutunda bir örneklem çizin:

from numpy import random

x = random.chisquare(df=2, size=(2, 3))

print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.chisquare(df=1, size=1000), kind="kde")

plt.show()
```

# **Zipf dağılımları,**

 Zipf yasasına dayalı olarak veri örneklemesi yapmak için kullanılır.

Zipf Yasası: Bir kelime grubunda, n. en sık kullanılan terim, en sık kullanılan terimin 1/n katı kadar sıklıkta geçer. Örneğin, İngilizcede en sık kullanılan 5. kelime, en sık kullanılan kelimeye göre yaklaşık 1/5 kat daha sık geçer.

İki parametresi vardır:

a- dağıtım parametresi.

size- Döndürülen dizinin şekli.

```python
#Zipf dağılımı için, dağılım parametresi 2 olan ve 2x3 boyutunda bir örneklem çizin:

from numpy import random

x = random.zipf(a=2, size=(2, 3))

print(x)
```

```python
#Daha anlamlı bir grafik elde etmek için 1000 noktadan örneklem alın, ancak yalnızca değeri < 10 olanları grafiğe ekleyin.
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.zipf(a=2, size=1000)
sns.displot(x[x<10])

plt.show()
```
