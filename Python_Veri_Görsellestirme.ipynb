# **Yoğunluk Grafikleri kdeplot()**

Yoğunluk grafikleri, diğer adıyla çekirdek yoğunluk grafikleri, sürekli bir değişkenin dağılımını gösteren bir veri görselleştirme türüdür. Histogramlara benzerler, ancak verileri çubuklar olarak temsil etmek yerine, yoğunluk grafikleri verilerin yoğunluğunu tahmin etmek için düzgün bir eğri kullanır. Seaborn'da yoğunluk grafikleri, fonksiyonu kullanılarak oluşturulabilir kdeplot(). Histogramın düzeltilmiş şeklidir diyebiliriz.

Çekirdek yoğunluk tahmini (KDE) grafiği, bir veri kümesindeki gözlemlerin dağılımını, histograma benzer şekilde görselleştirmek için kullanılan bir yöntemdir. KDE, verileri bir veya daha fazla boyutta sürekli bir olasılık yoğunluk eğrisi kullanarak temsil eder.

Histograma kıyasla, KDE özellikle birden fazla dağılım çizilirken daha az karmaşık ve daha yorumlanabilir bir grafik üretebilir. Ancak, altta yatan dağılım sınırlı veya düzgün değilse, bozulmalara yol açma potansiyeline sahiptir. Histogramda olduğu gibi, gösterimin kalitesi de iyi düzeltme parametrelerinin seçimine bağlıdır.

```python

sns.kdeplot(data=modely, x="prices")
```

```python
sns.kdeplot(data=modely, x="prices",shade=True)
```

```python
# Yoğunluk grafiği (KDE Plot) için ayarlamalar yapalım.
# sns.kdeplot(data=modely, x="prices")

# fill: Yoğunluk eğrisinin altını doldurur. (True/False)
# color: Çizginin rengini belirler. (örn: 'blue', 'red', 'green', hex kodu)
# linestyle: Çizginin stilini belirler. (örn: '-', '--', ':', '-.')
# linewidth: Çizginin kalınlığını belirler.
# alpha: Çizginin ve dolgunun şeffaflığını ayarlar (0.0 - 1.0 arası).
# label: Lejantta gösterilecek etiketi belirler.
# bw_adjust: Bant genişliğini ayarlayarak eğrinin düzgünlüğünü kontrol eder. Yüksek değer daha düzgün, düşük değer daha dalgalı.
# cut: Eğrinin veri aralığının ne kadar ötesine uzanacağını belirler.
# gridsize: İç tahmin için kullanılan ızgara noktası sayısını ayarlar.
# cumulative: Kümülatif yoğunluk fonksiyonu (CDF) çizer. (True/False)
# hue: Belirtilen kategorik değişkene göre farklı yoğunluk eğrileri çizer.
# common_norm: Eğer 'hue' kullanılıyorsa, tüm eğrilerin aynı yoğunluk ölçeğini paylaşmasını sağlar (True/False).

plt.figure(figsize=(9, 6))
sns.kdeplot(
    data=modely,
    x="prices",
    fill=True,       # Eğrinin altını doldur
    color="purple",  # Eğrinin rengi
    linewidth=2,     # Çizgi kalınlığı
    alpha=0.6,       # Şeffaflık
    label="Fiyat Yoğunluğu", # Lejant etiketi
    bw_adjust=0.8,   # Bant genişliği ayarı
    cut=0            # Eğrinin veri aralığının dışında ne kadar çizileceği
)

plt.title('Model Y Fiyat Dağılımı Yoğunluk Grafiği', fontsize=16)
plt.xlabel('Fiyatlar', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# 'hue' parametresiyle farklı model detaylarına göre yoğunlukları karşılaştırma:
plt.figure(figsize=(9, 6))
sns.kdeplot(
    data=modely,
    x="prices",
    hue="model_details", # 'model_details' sütununa göre farklı çizgiler çiz
    fill=False,#shade yerine kullanılır. shade devredışı kaldı.
    alpha=0.4,
    linewidth=2.5,
    common_norm=False # Her kategori için kendi normunu kullan
)
plt.title('Model Detaylarına Göre Fiyat Yoğunluk Dağılımı', fontsize=16)
plt.xlabel('Fiyatlar', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)
# plt.legend(modely.model_details,loc='best') # Bu hatalı çağrıyı kaldırıyoruz.
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9, 6))

sns.kdeplot(data=modely, x="km", fill=True)

plt.title('Km Yoğunluk Grafiği', fontsize=16)
plt.xlabel('Km', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
#Daha kısa yoldan şu şekilde de çizebiliriz.

sns.kdeplot(modely.prices)
#
```

```python
# Bant genişliği (bw_adjust) değeri, yoğunluk grafiğinin pürüzsüzlüğünü ayarlar.
# Küçük değerler daha 'dalgalı' bir grafik oluştururken, büyük değerler daha 'pürüzsüz' bir grafik oluşturur.
# Optimum değer, verinizin yapısına ve ne kadar detay görmek istediğinize bağlıdır.
# Genellikle 0.5 ile 2.0 arasında değerler denenerek görsel olarak en uygunu bulunur.

# bw_adjust=0.5 ile daha detaylı bir grafik
sns.kdeplot(modely.prices, bw_adjust=0.5, label='bw_adjust = 0.5')

# bw_adjust=1 (varsayılan civarı) ile orta düzeyde bir grafik
sns.kdeplot(modely.prices, bw_adjust=1.0, label='bw_adjust = 1.0')

# bw_adjust=2 ile daha pürüzsüz bir grafik
sns.kdeplot(modely.prices, bw_adjust=2.0, label='bw_adjust = 2.0')

plt.title('Fiyat Dağılımı (Farklı Bant Genişlikleri)')
plt.xlabel('Fiyatlar')
plt.ylabel('Yoğunluk')
plt.legend()
plt.show()
```

```python
#Bu parametre, KDE grafiğinin kümülatif olarak çizilmesini sağlar. Normal bir KDE, bir değerin olasılık yoğunluğunu gösterirken,
#cumulative=True olduğunda, belirli bir değere kadar olan gözlemlerin yüzdesini gösterir. Yani, grafikteki herhangi bir X noktasında Y değeri,
#tüm verilerin Y kadarının X'ten küçük veya X'e eşit olduğu anlamına gelir.

sns.kdeplot(modely.prices,cumulative=True)
```

```python
#iki boyutlu yoğunluk grafiği oluşturmak istersek
sns.kdeplot(x=modely.prices, y=modely.km, fill=True)
plt.title('Fiyat ve KM Arası İki Boyutlu Yoğunluk Dağılımı')
plt.xlabel('Fiyatlar')
plt.ylabel('Kilometre')
plt.show()
```

```python
#iki boyutlu yoğunluk grafiği oluşturmak istersek

# sns.kdeplot(): İki boyutlu (bivariate) yoğunluk tahmini grafiği çizer.
sns.kdeplot(
    x=modely.prices, # X ekseni için 'prices' sütununu kullanır.
    y=modely.km,      # Y ekseni için 'km' sütununu kullanır.
    fill=True,        # Yoğunluk alanlarının içini doldurur.
    shade_lowest=False, # En düşük yoğunluk alanını gösterme
    cmap='viridis',   # Doldurulan alanlar için 'viridis' renk haritasını kullanır.
    levels=10,        # Yoğunluk dağılımını gösteren 10 farklı kontur seviyesi çizer.
    cbar=True         # Grafiğin sağ tarafına renklerin yoğunluk seviyelerini gösteren bir renk çubuğu (colorbar) ekler.
)
# plt.title(): Grafiğe bir başlık ekler.
plt.title('Fiyat ve KM Arası İki Boyutlu Yoğunluk Dağılımı')
# plt.xlabel(): X eksenine etiket ekler.
plt.xlabel('Fiyatlar')
# plt.ylabel(): Y eksenine etiket ekler.
plt.ylabel('Kilometre')
# plt.show(): Oluşturulan grafiği görüntüler.
plt.show()
```

```python
# Yeni bir figür ve eksenler oluşturuyoruz.
plt.figure(figsize=(10, 6))

# İlk KDE grafiğini çiziyoruz (Fiyatlar için).
# x: X ekseni verisi.
# fill=True: Eğrinin altını doldurur.
# alpha: Dolgu renginin şeffaflığını ayarlar (0.0 tamamen şeffaf, 1.0 opak).
# label: Lejantta görünecek etiket.
sns.kdeplot(x=modely.prices, fill=True, alpha=0.5, label='Fiyat Yoğunluğu', color='blue')

# İkinci KDE grafiğini çiziyoruz (Kilometre için).
# Bu grafik ilk grafiğin üzerine çizilecektir.
sns.kdeplot(x=modely.km, fill=True, alpha=0.5, label='KM Yoğunluğu', color='red')

# Grafiğe başlık ve eksen etiketleri ekliyoruz.
plt.title('Fiyat ve KM Dağılımları (Üst Üste)', fontsize=16)
plt.xlabel('Değerler', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)

# Lejantı gösteriyoruz, böylece hangi eğrinin hangi veriye ait olduğu anlaşılır.
plt.legend()

# Izgarayı ekliyoruz.
plt.grid(True, linestyle='--', alpha=0.7)

# Grafiği düzenli gösteriyoruz.
plt.tight_layout()

# Grafiği ekrana basıyoruz.
plt.show()
```

```python
# Seaborn kütüphanesini kullanarak bir kutu grafiği (boxplot) oluşturur.
# Kutu grafiği, sayısal verilerin dağılımını, çeyrekleri ve aykırı değerleri görselleştirmek için kullanılır.
sns.boxplot(
    data=araba, # Kullanılacak DataFrame'i belirtiriz (önceden yüklenmiş 'araba' veri seti).
    x= araba['origin'], # X ekseninde 'origin' (menşe ülke) sütunundaki verileri kullanırız. Bu, kategorik bir değişkendir.
    y= araba['mpg'], # Y ekseninde 'mpg' (mil başına galon) sütunundaki verileri kullanırız. Bu, sayısal bir değişkendir.
    hue= araba['model_year']>76, # 'model_year' sütunundaki değerlerin 76'dan büyük olup olmadığına göre farklı renklerde kutular çizer.
                          # Bu, 1976 sonrası ve 1976 öncesi araçların yakıt tüketimini karşılaştırmamızı sağlar.
    palette="bright", # Grafikte kullanılacak renk paletini 'bright' olarak ayarlar. Daha canlı renkler sağlar.
    order=['europe','japan','usa'], # X eksenindeki kategorilerin (origin) gösterilme sırasını belirleriz.
    width=0.5,#kutuların genişliğini ayarlarız.
    linewidth=2, # kutuların kalınlığını ayarlarız.
    # whis: Bıyıkların uzunluğunu kontrol eder ve aykırı değerleri belirlemede kullanılır.
    # Varsayılan olarak 1.5, IQR'nin 1.5 katı dışındaki değerleri aykırı değer kabul eder.
    # Bu değeri düşürerek (örneğin 0.5) daha fazla değerin aykırı olarak gösterilmesini sağlayabilirsiniz.
    whis=1.5,
    # fliersize: Aykırı değerleri temsil eden noktaların boyutunu ayarlar.
    # Aykırı değerler, kutu grafiğinin bıyıklarının dışına düşen veri noktalarıdır.
    # Şu anki ayarlarla (whis=1.5), veri setinde bu kriterlere uyan bir aykırı değer bulunmadığı için grafikte nokta görünmez.
    fliersize= 3
    )
```

```python
#Aykırı değerlerin neler olduğunu grafikte kullanmak istersek şöyle bir kod kullanabiliriz.

import seaborn as sns
import matplotlib.pyplot as plt

# 1. Grafiği çiz ve 'ax' değişkenine ata
plt.figure(figsize=(12, 6))
ax = sns.boxplot(
    data=araba,
    x='origin', # Not: araba['origin'] yerine sadece sütun adı yazmak daha sağlıklıdır
    y='mpg'
)

# 2. Aykırı değerleri bulup üzerine numara yazan kısım
for i, group in enumerate(araba['origin'].unique()):
    # Her bir grubun verisini alalım
    subset = araba[araba['origin'] == group]['mpg']

    # IQR hesaplama
    q1 = subset.quantile(0.25)
    q3 = subset.quantile(0.75)
    iqr = q3 - q1

    # Alt ve üst sınır
    alt = q1 - 1.5 * iqr
    ust = q3 + 1.5 * iqr

    # Aykırıları filtrele
    outliers = subset[(subset < alt) | (subset > ust)]

    # Grafiğe ekle
    for index, value in outliers.items():
        ax.annotate(index,
                    xy=(i, value),
                    xytext=(5, 0),
                    textcoords='offset points',
                    color='red',
                    fontsize=10,
                    weight='bold')

plt.title("Origin'e Göre MPG ve Aykırı Değer İndeksleri")
plt.show()
```
