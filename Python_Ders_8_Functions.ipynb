# **Python Fonksiyonları**

Fonksiyon, yalnızca çağrıldığında çalışan bir kod bloğudur.

Bir fonksiyon sonuç olarak veri döndürebilir.

Fonksiyonlar, kod tekrarını önlemeye yardımcı olur.

**Python'da fonksiyon adları, değişken adlarıyla aynı kurallara tabidir: **

Bir fonksiyon adı bir harf veya alt çizgi ile başlamalıdır.

Bir fonksiyon adı yalnızca harf, rakam ve alt çizgi içerebilir.

Fonksiyon adları büyük/küçük harf duyarlıdır ( myFunctionve myfunctionfarklıdır).

```python
# def ile başlar.
#Fonksiyonun içindeki kod girintili olmalıdır. Python, kod bloklarını tanımlamak için girintiyi kullanır.
def my_function():
  print("Hello from a function")
```

```python
#Bir fonksiyonu çağırmak için, fonksiyonun adını parantez içinde yazın:
#Aynı fonksiyonu birden fazla kez çağırabilirsiniz:

def my_function():
  print("Hello from a function")

my_function()
my_function()
my_function()
```

```python
#Fonksiyonlar sayesinde - yeniden kullanılabilir kod:
def my_function(fname):
  print(fname + " Refsnes")

my_function("Emil")
my_function("Tobias")
my_function("Linus")
```

```python
#Değer döndüren bir fonksiyon:
def get_greeting():
  return "Hello from a function"

get_greeting()
```

```python
#Fonksiyon tanımları boş olamaz. Herhangi bir kod içermeyen bir fonksiyon yer tutucusu oluşturmanız gerekiyorsa, şu passifadeyi kullanın:
def my_function():
  pass
```

```python
#Argüman , bir fonksiyon çağrıldığında ona gönderilen gerçek değerdir.
#Parametre , fonksiyon tanımında parantez içinde belirtilen değişkendir.

def my_function(name): # name is a parameter
  print("Hello", name)

my_function("Emil") # "Emil" is an argument
```

```python
#Varsayılan olarak, bir fonksiyon doğru sayıda argümanla çağrılmalıdır.
#Eğer fonksiyonunuz 2 argüman bekliyorsa, onu tam olarak 2 argümanla çağırmalısınız.
def my_function(fname, lname):
  print(fname + " " + lname)

my_function("Emil", "Refsnes")
```

```python
#Parametrelere varsayılan değerler atayabilirsiniz. Fonksiyon argüman olmadan çağrıldığında varsayılan değeri kullanır:
def my_function(country = "Norway"):
  print("I am from " + country)

my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
```

```python
#Anahtar = değer sözdizimini kullanarak argümanlar gönderebilirsiniz .
def my_function(animal, name):
  print("I have a", animal)
  print("My", animal + "'s name is", name)

my_function(animal = "dog", name = "Buddy")
```

```python
#Bu şekilde, anahtar kelime argümanlarıyla, argümanların sırası önemli değildir.
#"Keyword Arguments" ifadesi Python dokümanlarında genellikle "kwargs" olarak kısaltılır .

def my_function(animal, name):
  print("I have a", animal)
  print("My", animal + "'s name is", name)

my_function(name = "Buddy", animal = "dog")
```

```python
#Fonksiyonları anahtar kelimeler kullanmadan argümanlarla çağırdığınızda, bunlara konumlu argümanlar denir.
#Konumsal argümanlar doğru sırada olmalıdır:
#Konumsal argümanlarda sıralama önemlidir:

def my_function(animal, name):
  print("I have a", animal)
  print("My", animal + "'s name is", name)

my_function("dog", "Buddy")
```

```python
#Bir fonksiyona argüman olarak herhangi bir veri türünü (dize, sayı, liste, sözlük vb.) gönderebilirsiniz.
#Fonksiyon içinde veri türü korunacaktır:
```

```python
def my_function(fruits):
  for fruit in fruits:
    print(fruit)

my_fruits = ["apple", "banana", "cherry"]
my_function(my_fruits)
```

```python
#Argüman olarak sözlük gönderme:
def my_function(person):
  print("Name:", person["name"])
  print("Age:", person["age"])

my_person = {"name": "Emil", "age": 25}
my_function(my_person)
```

```python
#Fonksiyonlar şu ifadeyi kullanarak (return) değer döndürebilir :
#Fonksiyonlar, listeler, demetler, sözlükler ve daha fazlası dahil olmak üzere her türlü veri türünü döndürebilir.
  return ["apple", "banana", "cherry"]

fruits = my_function()
print(fruits[0])
print(fruits[1])
print(fruits[2])
```

```python
def my_function(x, y):
  return x + y

result = my_function(5, 3)
print(result)
```

```python
#Bir demet (tuple) döndüren bir fonksiyon:
def my_function():
  return (10, 20)

x, y = my_function()
print("x:", x)
print("y:", y)
```

```python
#Bir fonksiyonun yalnızca anahtar kelime argümanları alabileceğini belirtmek için, argümanların *, önüne şunu ekleyin:
def my_function(*, name):
  print("Hello", name)

my_function(name = "Emil")
```

```python
#İki argüman türünü aynı fonksiyonda birleştirebilirsiniz.
#Önceki argümanlar /yalnızca konum tabanlıdır, sonraki argümanlar ise *yalnızca anahtar kelime tabanlıdır:

def my_function(a, b, /, *, c, d):
  return a + b + c + d

result = my_function(5, 10, c = 15, d = 20)
print(result)
```

```python
#Varsayılan olarak, bir fonksiyon doğru sayıda argümanla çağrılmalıdır.

#Ancak bazen fonksiyonunuza kaç argümanın iletileceğini bilemeyebilirsiniz.

#*args ve **kwargs, fonksiyonların bilinmeyen sayıda argüman kabul etmesine olanak tanır.

#Fonksiyonunuza kaç argüman geçirileceğini bilmiyorsanız, *parametre adının önüne bir `<argüman adı>` ekleyin.

#Bu şekilde, fonksiyon bir argüman demeti alacak ve buna göre öğelere erişebilecektir:

def my_function(*kids):
  print("The youngest child is " + kids[2])

my_function("Emil", "Tobias", "Linus")
```

```python
#Bu *argsparametre, bir fonksiyonun herhangi bir sayıda konumlu argümanı kabul etmesine olanak tanır.

#Fonksiyonun içinde, argsgeçirilen tüm argümanları içeren bir demet (tuple) oluşur:

def my_function(*args):
  print("Type:", type(args))
  print("First argument:", args[0])
  print("Second argument:", args[1])
  print("All arguments:", args)

my_function("Emil", "Tobias", "Linus")
```

```python
#Normal parametreleri *args ile birleştirebilirsiniz.

#Normal parametreler *args'tan önce gelmelidir:

def my_function(greeting, *names):
  for name in names:
    print(greeting, name)

my_function("Hello", "Emil", "Tobias", "Linus")
```

```python
#Herhangi bir sayıda değerin toplamını hesaplayan bir fonksiyon:

def my_function(*numbers):
  total = 0
  for num in numbers:
    total += num
  return total

print(my_function(1, 2, 3))
print(my_function(10, 20, 30, 40))
print(my_function(5))
```

```python
#Maksimum değeri bulma:
def my_function(*numbers):
  if len(numbers) == 0:
    return None
  max_num = numbers[0]
  for num in numbers:
    if num > max_num:
      max_num = num
  return max_num

print(my_function(3, 7, 2, 9, 1))
```

```python
#Fonksiyonunuza kaç adet anahtar kelime argümanı geçirileceğini bilmiyorsanız, **parametre adının önüne iki yıldız işareti ekleyin.

#Bu şekilde, fonksiyon bir argüman sözlüğü alacak ve buna göre öğelere erişebilecektir:

#**kwargs kullanarak istenilen sayıda anahtar kelime argümanı kabul edebilirsiniz:

def my_function(**person):
  print("Name:", person["name"])
  print("Age:", person["age"])

my_function(name = "Emil", age = 25)
```

```python
#Bu **kwargsparametre, bir fonksiyonun herhangi bir sayıda anahtar kelime argümanı kabul etmesine olanak tanır.

#Fonksiyonun içinde, kwargstüm anahtar kelime argümanlarını içeren bir sözlük oluşur:

def my_function(**myvar):
  print("Type:", type(myvar))
  print("Name:", myvar["name"])
  print("Age:", myvar["age"])
  print("All data:", myvar)

my_function(name = "Tobias", age = 30, city = "Bergen")
```

```python
#Normal parametreleri **kwargs** ile birleştirebilirsiniz.

#Normal parametreler **kwargs**'tan önce gelmelidir:

def my_function(greeting, **names):
  for name in names:
    print(greeting, names[name])

my_function("Hello", first_name = "Emil", last_name = "Refsnes")

```

```python
def my_function(title, *args, **kwargs):
  print("Title:", title)
  print("Positional arguments:", args)
  print("Keyword arguments:", kwargs)

my_function("User Info", "Emil", "Tobias", age = 25, city = "Oslo")
```

```python
#* kullanarak bir listeyi argümanlara açma:

def my_function(a, b, c):
  return a + b + c

numbers = [1, 2, 3]
result = my_function(*numbers) # Same as: my_function(1, 2, 3)
print(result)
```

```python
#** operatörünü kullanarak bir sözlüğü anahtar kelime argümanlarına açma:

def my_function(fname, lname):
  print("Hello", fname, lname)

person = {"fname": "Emil", "lname": "Refsnes"}
my_function(**person) # Same as: my_function(fname="Emil", lname="Refsnes")
```

```python
#Bir fonksiyon içinde oluşturulan bir değişken, o fonksiyonun yerel kapsamına aittir ve yalnızca o fonksiyon içinde kullanılabilir.

def myfunc():
  x = 300
  print(x)

myfunc()
```

```python
#Yerel değişkene, bir fonksiyonun içindeki başka bir fonksiyondan erişilebilir:

def myfunc():
  x = 300
  def myinnerfunc():
    print(x)
  myinnerfunc()

myfunc()
```

```python
#Bir fonksiyonun dışında oluşturulan bir değişken globaldir ve herkes tarafından kullanılabilir:
x = 300

def myfunc():
  print(x)

myfunc()

print(x)
```

```python
#Fonksiyon önce yerel değişkeni x, ardından kod genel değişkeni yazdıracaktır x:
x = 300

def myfunc():
  x = 200
  print(x)

myfunc()

print(x)
```

```python
#Eğer global bir değişken oluşturmanız gerekiyorsa ancak yerel kapsamda sıkışıp kaldıysanız, `<global>` anahtar kelimesini kullanabilirsiniz global.

#Bu globalanahtar kelime değişkeni global hale getirir.

def myfunc():
  global x
  x = 300

myfunc()

print(x)
```

```python
x = 300

def myfunc():
  global x
  x = 200

myfunc()

print(x)
```

```python
#Dekoratörler, bir fonksiyonun kodunu değiştirmeden ona ekstra davranışlar eklemenizi sağlar.

#Dekoratör, girdi olarak başka bir fonksiyon alan ve yeni bir fonksiyon döndüren bir fonksiyondur.

#Fonksiyon tanımının hemen üstüne yerleştirilerek @changecase, fonksiyon myfunctionfonksiyonla "süslenmiş" oluyor changecase .

#Fonksiyon, changecasedekoratördür.

#myfunctionSüsleme yapılacak olan fonksiyon budur .

def changecase(func):
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Hello Sally"

print(myfunction())
```

```python
#@changecaseDekoratörü iki fonksiyonda kullanma :
def changecase(func):
  def myinner():
    return func().upper()
  return myinner

@changecase
def myfunction():
  return "Hello Sally"

@changecase
def otherfunction():
  return "I am speed!"

print(myfunction())
print(otherfunction())
```

```python
#Argüman alan fonksiyonlar da süslenebilir:

def changecase(func):
  def myinner(x):
    return func(x).upper()
  return myinner

@changecase
def myfunction(nam):
  return "Hello " + nam

print(myfunction("John"))
```

```python
#Normalde, bir fonksiyonun adı şu öznitelikle birlikte döndürülebilir __name__:
def myfunction():
  return "Have a great day!"

print(myfunction.__name__)
```

```python
#Lambda fonksiyonu, küçük, anonim bir fonksiyondur.
#Lambda fonksiyonları herhangi bir sayıda argüman alabilir, ancak yalnızca bir ifade içerebilir.
x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a, b : a * b
print(x(5, 6))
```

```python
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

```python
#Lambda fonksiyonlarının gücü, onları başka bir fonksiyonun içinde anonim bir fonksiyon olarak kullandığınızda daha iyi ortaya çıkar.
#Diyelim ki, tek bir argüman alan ve bu argümanın bilinmeyen bir sayıyla çarpılacağı bir fonksiyon tanımınız var:
#Bu fonksiyon tanımını kullanarak, gönderdiğiniz sayıyı her zaman ikiye katlayan bir fonksiyon oluşturun:

def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11))
print(mytripler(11))
```

```python
#Listedeki tüm sayıları ikiye katlayın:
numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)
```

```python
#Listedeki çift sayıları filtrele:
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))
print(odd_numbers)
```

```python
#Bir demet listesini ikinci elemanına göre sıralayın:
students = [("Emil", 25), ("Tobias", 22), ("Linus", 28)]
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
```

```python
#Dizeleri uzunluğa göre sırala:
strings = ["apple", "banana", "cherry", "date"]
sorted_strings = sorted(strings, key=lambda x: len(x))
print(sorted_strings)
```
