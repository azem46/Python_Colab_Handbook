# **R'IN KENDİ BÜNYESİNDE YER ALAN PAKETLERLE GRAFİK OLUŞTURMA**

```r
#Basit Bir Grafik oluşturma
a = c(12,25,30,32,25)
b = c(20,15,20,24,35)

plot(a,b,
     xlab = "X Ekseni Başlık",
     ylab = "Y Ekseni Başlık"
     )
text(20,25,"Berin") #girilen koordinata isim verdik
abline(h = 18) #grafiğe yatar çizgi ekledik, dikey için v değeri girebiliriz.
lines(c(15,20),c(30,35)) #istenilen bir kısıma doğru çizdirdik
```

```r
#R ın kendi bünyesinde bulunan datalardan olan kadın women veri siteni çağırıyoruz. Ve Düzenleme yapıyoruz.
datasets::women
kadin=women
colnames(kadin) = c("Ağırlık","uzunluk")
kadin
```

```r
#Kadınların evli olup olamdığıyla ilgili datamız geldi onu mevcud datamıza ekliyoruz.
evli_mi=c(T,T,F,F,F,T,F,F,T,T,F,F,F,F,T)

kadin1 = cbind(kadin,evli_mi)
kadin1
```

```r
#Nokta grafiğini çizdirelim

plot(kadin1$`Ağırlık`,kadin1$uzunluk,
     main = "Boy Kilo Grafiği",
     xlab = "Ağırlık",
     ylab = "Uzunluk",
     type ="o"#noktaları çizgi iel birleştirdik
     )
```

```r
#Ağırlığa ait histogram grafiği

hist(kadin1$uzunluk,main = "Uzunluk Dağılımı")
```

```r
#kutu grafiğini çizdirmek için

boxplot(kadin1$uzunluk,main="Kutu Grafiği")
```

```r
#Paketler hakkında bilgi almak için paket isminin başına ? karakteri bırakmamız ysterli

?plot
```

```r
#Pasta Grafik

pie(table(kadin1$evli_mi),main = "Medeni Durum Grafiği")
```
