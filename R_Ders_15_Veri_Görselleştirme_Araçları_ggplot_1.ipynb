# **R'IN KENDİ BÜNYESİNDE YER ALAN PAKETLERLE GRAFİK OLUŞTURMA**

```r
#Basit Bir Grafik oluşturma
a = c(12,25,30,32,25)
b = c(20,15,20,24,35)

plot(a,b,
     xlab = "X Ekseni Başlık",
     ylab = "Y Ekseni Başlık"
     )
text(20,25,"Berin") #girilen koordinata isim verdik
abline(h = 18) #grafiğe yatar çizgi ekledik, dikey için v değeri girebiliriz.
lines(c(15,20),c(30,35)) #istenilen bir kısıma doğru çizdirdik
```

```r
#R ın kendi bünyesinde bulunan datalardan olan kadın women veri siteni çağırıyoruz. Ve Düzenleme yapıyoruz.
datasets::women
kadin=women
colnames(kadin) = c("Ağırlık","uzunluk")
kadin
```

```r
#Kadınların evli olup olamdığıyla ilgili datamız geldi onu mevcud datamıza ekliyoruz.
evli_mi=c(T,T,F,F,F,T,F,F,T,T,F,F,F,F,T)

kadin1 = cbind(kadin,evli_mi)
kadin1
```

```r
#Nokta grafiğini çizdirelim

plot(kadin1$`Ağırlık`,kadin1$uzunluk,
     main = "Boy Kilo Grafiği",
     xlab = "Ağırlık",
     ylab = "Uzunluk",
     type ="o"#noktaları çizgi iel birleştirdik
     )
```

```r
#Ağırlığa ait histogram grafiği

hist(kadin1$uzunluk,main = "Uzunluk Dağılımı")
```

```r
#kutu grafiğini çizdirmek için

boxplot(kadin1$uzunluk,main="Kutu Grafiği")
```

```r
#Paketler hakkında bilgi almak için paket isminin başına ? karakteri bırakmamız ysterli

?plot
```

```r
#Pasta Grafik

pie(table(kadin1$evli_mi),main = "Medeni Durum Grafiği")
```

# **GGPLOT İLE VERİ GÖRSELLEŞTİRME**

```r
#paketin yüklenmesi ve eklenmesi gerekiyor
#install.packages(ggplot2) tidyversle birlikte geliyor
library(ggplot2)
#Grafikler katmanlardan oluşur ve aşama aşama oluşturulur
```

```r
arabalar= datasets::mtcars # arabalara ait data setimizi çağırdık ve bunu arabalar değişkenine atadık.
View(arabalar) #oluşturduğumuz bu datayı ayrı bir pencerede açtık.





```

```r
#arabalar datasından küçük bir data oluşturuyoruz.
araba_veri = subset(arabalar,select = c("mpg","cyl","wt","am"))
```

```r
#oluşturduğumuz bu data üzerinde düzeltme yapıyoruz.

colnames(araba_veri)=c("yol","silindir","ağırlık","vites")
```

```r
#Değişkenlerin bazısını kategorik Değişkene Çeviriyoruz.

araba_veri$silindir = as.factor(araba_veri$silindir)
araba_veri$vites = as.factor(araba_veri$vites)
levels(araba_veri$vites)= c("otomatik","manuel") #kategork değişkenin alt kategorilerine isimler verdik.
```

```r
#oluşturduğumuz yeni veri tablomuz
#Bundan sonraki örneklerde bu veriyi kullanacağız.
View(araba_veri)
```

# **Systax Dizilimi**
ggplot(Veri Seti,aes(x ekseni için değişken, y ekseni için değişken))

```r
#Grafiğin temelini atıyoruz.

grafik1 = ggplot(araba_veri,aes(ağırlık, yol)) +

# şimdide geometrik şekilleri ekliyoruz

geom_point()# nokta grafiği ekledik

grafik1 #grafiği yazdırdık
```

```r
grafik2 = ggplot(araba_veri,aes(silindir,yol))+
geom_point(
colour="red",#noktaları renklendirdik.
shape=17,#17 koduyla noktayı üçgene dönüştürdük
size =4 #üçgen boyutunu ayarladık.
)
grafik2
```

```r
# grafiğe yeni bir katman ekliyoruz vites türü değişik renklerde gelecek

grafik3= ggplot(araba_veri,
                aes(silindir,yol ))+ #her yeni katman + sembolü ile eklenir
  geom_point(aes(
  colour= vites,
  size =5
  ))
grafik3
```

```r
# grafiğe yeni bir katman ekliyoruz noktalara isimler gelecek

grafik3= ggplot(araba_veri,
                aes(silindir,yol, label= row.names(araba_veri)))+ #her yeni katman + sembolü ile eklenir
  geom_point(aes(
  colour= vites,
  size =5
  ))+
  geom_text(check_overlap = TRUE)# check ayarı grafiği daha okunaklı hale getirdi

grafik3
```

```r
#Saçılım grafiğinde çalışan noktaların yayarak görmek için

grafik3= ggplot(araba_veri,
                aes(silindir,yol ))+ #her yeni katman + sembolü ile eklenir
  geom_point(aes(
  colour= vites,
  size =5,
  shape= vites #alt kategorileri değişik sembollerle göstermek için
  ),
  position = "jitter")#bu kodla üst üste binen noktaları yaydık
grafik3
```
