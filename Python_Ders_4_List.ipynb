# **Listeler**

Listeler, birden fazla öğeyi tek bir değişkende saklamak için kullanılır.

Listeler, Python'da veri koleksiyonlarını depolamak için kullanılan 4 yerleşik veri türünden biridir; diğer 3'ü ise Tuple , Set ve Dictionary'dir ve her birinin farklı özellikleri ve kullanım alanları vardır.

Listeler köşeli parantezler kullanılarak oluşturulur.

Liste öğeleri sıralıdır, değiştirilebilir ve aynı değerleri tekrar tekrar içerebilir.

Liste öğeleri indekslenmiştir; ilk öğenin indeksi [0], ikinci öğenin indeksi [1]vb. şeklindedir.

Liste değiştirilebilir, yani oluşturulduktan sonra listedeki öğeleri değiştirebilir, ekleyebilir ve kaldırabiliriz.

Listeler indekslendiği için, listelerde aynı değere sahip öğeler
bulunabilir.

```python
#Bir listenin kaç öğe içerdiğini belirlemek için şu len()işlevi kullanılır.
thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

```python
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
list4 = ["abc", 34, True, 40, "male"]

print(list1)
print(list2)
print(list3)
print(list4)
```

```python
#Type kontrolü için
print(type(list1))
```

```python
#List yapıcı list() Fonksiyonu, Tuple veri tipinin list veri tipine dönüşmesi
thislist = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)
```

```python
#Liste öğelerine erişim
thislist = ["apple", "banana", "cherry"]
print(thislist[1])
```
