# **Listeler**

Listeler, birden fazla öğeyi tek bir değişkende saklamak için kullanılır.

Listeler, Python'da veri koleksiyonlarını depolamak için kullanılan 4 yerleşik veri türünden biridir; diğer 3'ü ise Tuple , Set ve Dictionary'dir ve her birinin farklı özellikleri ve kullanım alanları vardır.

Listeler köşeli parantezler kullanılarak oluşturulur.

Liste öğeleri sıralıdır, değiştirilebilir ve aynı değerleri tekrar tekrar içerebilir.

Liste öğeleri indekslenmiştir; ilk öğenin indeksi [0], ikinci öğenin indeksi [1]vb. şeklindedir.

Liste değiştirilebilir, yani oluşturulduktan sonra listedeki öğeleri değiştirebilir, ekleyebilir ve kaldırabiliriz.

Listeler indekslendiği için, listelerde aynı değere sahip öğeler
bulunabilir.

```python
#Bir listenin kaç öğe içerdiğini belirlemek için şu len()işlevi kullanılır.
thislist = ["apple", "banana", "cherry"]
print(len(thislist))
```

```python
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
list4 = ["abc", 34, True, 40, "male"]

print(list1)
print(list2)
print(list3)
print(list4)
```

```python
#Type kontrolü için
print(type(list1))
```

```python
#List yapıcı list() Fonksiyonu, Tuple veri tipinin list veri tipine dönüşmesi
thislist = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)
```

```python
#Liste öğelerine erişim
#Not: İlk öğenin indeksi 0'dır.
thislist = ["apple", "banana", "cherry"]
print(thislist[1])
```

```python
#Negatif indeksleme, sondan başlamak anlamına gelir.
#-1" Son öğeyi", " -2 sondan ikinci öğeyi" vb. ifade eder.
thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
```

```python
#Endeks Aralığı
#Başlangıç ​​ve bitiş noktalarını belirterek bir indeks aralığı tanımlayabilirsiniz.
#Bir aralık belirtildiğinde, dönüş değeri belirtilen öğeleri içeren yeni bir liste olacaktır.
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
```

```python
#Bu örnek, başlangıçtan "kiwi"ye kadar olan ancak "kiwi"yi içermeyen öğeleri döndürür
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:4])
```

```python
#Bu örnek, "cherry"den sona kadar olan öğeleri döndürür.
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
```

```python
#Listede "elma" kelimesinin olup olmadığını kontrol edin:
thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
  print("Yes, 'apple' is in the fruits list")
```

```python
#Belirli bir öğenin değerini değiştirmek için, dizin numarasını kullanırız.
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
```

```python
#"Banana" ve "cherry" değerlerini "blackcurrant" ve "watermelon" değerleriyle değiştirir.
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)
```

```python
#Listenin 1. elemanı yerine iki yeni değer ekledik.
#Not: Eklenen öğe sayısı, değiştirilen öğe sayısıyla eşleşmediğinde listenin uzunluğu değişecektir.
thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant", "watermelon"]
print(thislist)
```

```python
#İkinci ve üçüncü değerleri tek bir değerle değiştirerek düzenler.
thislist = ["apple", "banana", "cherry"]
thislist[1:3] = ["watermelon"]
print(thislist)
```

```python
#Mevcut değerlerden herhangi birini değiştirmeden yeni bir liste öğesi eklemek için bu yöntemi kullanabiliriz insert().
thislist = ["apple", "banana", "cherry"]
thislist.insert(2, "watermelon")
print(thislist)
```

```python
#Listeye öğe eklemek için şu append() yöntemi kullanılır. Sona ekler.
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

```python
#Belirtilen dizine bir liste öğesi eklemek için bu insert()yöntemi kullanın.
thislist = ["apple", "banana", "cherry"]
thislist.insert(1, "orange")
print(thislist)
```

```python
#Başka bir listedeki öğeleri mevcut listeye eklemek için bu extend()yöntemi kullanın.
#Öğeler listenin sonuna eklenecektir .
thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)
```

```python
#Bu yöntem listeextend() eklemek zorunda değildir , herhangi bir yinelenebilir nesne (tuple, küme, sözlük vb.) ekleyebiliriz.
thislist = ["apple", "banana", "cherry"]
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(thislist)
```

```python
#Bu remove()yöntem, belirtilen öğeyi kaldırır.
thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)
```

```python
#Bu pop()yöntem, belirtilen dizini siler.
thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(thislist)
```

```python
#Bu delanahtar kelime ayrıca belirtilen dizini de siler.
thislist = ["apple", "banana", "cherry"]
del thislist[0]
print(thislist)
```

```python
#Bu delanahtar kelime listeyi tamamen silebilir de.
thislist = ["apple", "banana", "cherry"]
del thislist
```

```python
#Bu clear()yöntem listeyi boşaltır.
thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(thislist)
```

```python
#Listedeki tüm öğeleri tek tek yazdırır.
thislist = ["apple", "banana", "cherry"]
for x in thislist:
  print(x)
```

```python
thislist = ["apple", "banana", "cherry"]
for i in range(len(thislist)):
  print(thislist[i])
```

```python
thislist = ["apple", "banana", "cherry"]
i = 0
while i < len(thislist):
  print(thislist[i])
  i = i + 1
```

```python
#Bu range()fonksiyonu kullanarak yinelenebilir bir nesne oluşturabiliriz.
newlist = [x for x in range(10)]
print(newlist)
```

```python
#Yalnızca 5'ten küçük sayıları kabul eder.
newlist = [x for x in range(10) if x < 5]
print(newlist)
```

```python
#Liste nesnelerinin, sort()varsayılan olarak listeyi alfabetik ve sayısal olarak artan sırada sıralayan bir yöntemdir.
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort()
print(thislist)
```

```python
thislist = [100, 50, 65, 82, 23]
thislist.sort()
print(thislist)
```

```python
#Listeyi azalan sırada sıralamak için;
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort(reverse = True)
print(thislist)
```

```python
thislist = [100, 50, 65, 82, 23]
thislist.sort(reverse = True)
print(thislist)
```

```python
#büyük/küçük harf duyarlılığı olmayan bir sıralama fonksiyonu istiyorsanız, anahtar fonksiyon olarak `str.lower` kullanın:
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)
```

```python
#Bu reverse()yöntem, öğelerin mevcut sıralama düzenini tersine çevirir.
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

```python
#copy()Liste kopyalamak için yerleşik List yöntemini kullanabilirsiniz .
thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
```

```python
#Kopya oluşturmanın bir başka yolu da yerleşik yöntemi kullanmaktır list().
thislist = ["apple", "banana", "cherry"]
mylist = list(thislist)
print(mylist)
```

```python
#(:)Operatörü kullanarak bir listenin kopyasını oluşturur.
thislist = ["apple", "banana", "cherry"]
mylist = thislist[:]
print(mylist)
```

```python
#Python'da iki veya daha fazla listeyi birleştirmenin veya ek liste oluşturmanın çeşitli yolları vardır.
#En kolay yollardan biri  + operatörü kullanmaktır
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
```
