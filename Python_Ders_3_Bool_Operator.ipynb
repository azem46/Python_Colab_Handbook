# **Python Operatörler**

Aritmetik Operatörler

```python
x = 15
y = 4

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y) #Kalanı verir.
print(x ** y)# 15 in 4. Kuvvetini Verir.
print(x // y)# 15 in 4 e bölümünde bölümü verir.
```

 **Atama Operatörleri**

```python
# =
x = 5
```

```python
# +=Bu kod, x değişkenine 3 ekler ve sonucu tekrar x'e atar.
x += 3
print(x)
x = x + 3
print(x)
```

```python
#-=Bu kod, x değişkeninin değerinden 3 çıkarır ve sonucu tekrar x'e atar.
x = 12
x -= 3
print(x)
x = x - 3
print(x)
```

```python
#*=Bu kod, x değişkeninin değerini 3'le çarpar ve sonucu tekrar x'e atar.
x = 25
x *= 3
print(x)
x = x * 3
print(x)
```

```python
#/=Bu kod, x değişkeninin değerini 3'e böler ve sonucu tekrar x'e atar.
x = 15
x /= 3
print(x)
x = x / 3
print(x)
```

```python
#%= Bu kod, x değişkenine x'in 3'e bölümünden kalanı atar.
x = 15
x %= 3
print(x)
x = x % 3
print(x)
```

```python
#//=Bu kod, x değişkenine x'in 3'e tam sayı bölümünü atar.
x=20
x //= 3
print(x)
x = x // 3
print(x)
```

```python
#**=Bu kod, x değişkeninin 3. kuvvetini alıp sonucu tekrar x'e atar.
x=2
x **= 3
print(x)
x = x ** 3
print(x)
```

```python
#:=Bu kod, 'walrus operatörü' olarak bilinen := atama ifadesini kullanarak bir değişkeni hem atar hem de değerini döndürür.
x = 10
print(x := 3)
```

```python
numbers = [1, 2, 3, 4, 5]
count = len(numbers)
if count > 3:
    print(f"List has {count} elements")

if (count := len(numbers)) > 3:
    print(f"List has {count} elements")
```

**Karşılaştırma Operatörleri**

```python
x = 5
y = 3

print(x == y)#Eşit mi?
print(x != y)#Eşit Değil Mi?
print(x > y) #Büyük Mü?
print(x < y) # Küçük Mü?
print(x >= y) #Büyük Eşit Mi?
print(x <= y) #Küçük Eşit Mi?
```

**Kimlik Operatörleri**

(is ve is not) aynı mı sorgusu için True False Değer Döner

```python
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z)
print(x is y)
print(x == y)
```

```python
#x ve y değişkenlerinin bellekte aynı nesneye referans verip vermediğini kontrol eder.
x = ["apple", "banana"]
y = ["apple", "banana"]

print(x is not y)
```

```python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)
print(x is y)
```

**Üyelik Operatörleri**

in ve not in içinden mi kontrolü yapar varsa True döner.

```python
fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)
```

```python
#Yokluk kontrolü
fruits = ["apple", "banana", "cherry"]

print("pineapple" not in fruits)
```

```python
#Büyük küçük harf duyarlıdır.
text = "Hello World"

print("H" in text)
print("hello" in text)
print("z" not in text)
```
