# **Python Döngüleri**

*Python'da iki temel döngü komutu vardır:*

while döngüleri

for döngüleri

```python
#while Döngüsü
#Not: i'yi artırmayı unutmayın, aksi takdirde döngü sonsuza kadar devam eder.
while i < 6:
  print(i)
  i += 1
```

```python
#i değeri 3 olduğunda döngüden çık:
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
#i değeri 3 ise bir sonraki yinelemeye geçin:
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
#Koşul yanlış olduğunda bir mesaj yazdırın:
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```python
#Python For Döngüleri

#For döngüsü , bir diziyi (liste, demet, sözlük, küme veya dize olabilir) yinelemek için kullanılır.

#Bu, diğer programlama dillerindeki `for` anahtar kelimesine pek benzemez ve daha çok diğer nesne yönelimli programlama dillerinde bulunan yineleyici yöntemlere benzer şekilde çalışır.

#For döngüsü ile bir liste, demet, küme vb. içindeki her öğe için bir kez olmak üzere bir dizi ifadeyi çalıştırabiliriz.

#Meyve listesindeki her meyveyi yazdırın:

fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
#x"Banana" olduğunda döngüden çık :
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
#"Banana" olduğunda döngüden çık x, ancak bu sefer kesme işlemi yazdırma işleminden önce gerçekleşsin:
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

```python
#`continue` ifadesiyle döngünün mevcut yinelemesini durdurabilir ve bir sonrakine devam edebiliriz:

fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

```python
#Bir kod bloğunu belirtilen sayıda tekrarlamak için şu range()fonksiyonu kullanabiliriz:

#Bu range()fonksiyon, varsayılan olarak 0'dan başlayıp 1 artırarak (varsayılan olarak) belirtilen bir sayıda biten bir sayı dizisi döndürür.
#Dikkat edin, burada range(6)0 ile 6 arasındaki değerler değil, 0 ile 5 arasındaki değerler kastediliyor.

for x in range(6):
  print(x)
```

```python
#Başlangıç ​​parametresini kullanarak:
for x in range(2, 6):
  print(x)
```

```python
#Sırayı 3 artırın (varsayılan değer 1'dir):
for x in range(2, 30, 3):
  print(x)
```

```python
#0'dan 5'e kadar tüm sayıları yazdırın ve döngü bittiğinde bir mesaj yazdırın:
for x in range(6):
  print(x)
else:
  print("Finally finished!")
```

```python
#Değer 3 olduğunda döngüyü kırın xve bloğun ne olduğuna bakın
for x in range(6):
  if x == 3: break
  print(x)
else:
  print("Finally finished!")
```

```python
#İç içe döngü, bir döngünün içinde başka bir döngüdür.

#"İç döngü", "dış döngünün" her yinelemesi için bir kez çalıştırılacaktır:

adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

```python
#Döngüler boş olamaz, ancak herhangi bir nedenle içeriği olmayan bir döngünüz varsa , hata almamak için forbu ifadeyi ekleyin .pass
for x in [0, 1, 2]:
  pass
```
